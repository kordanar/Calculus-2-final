```python
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d.axes3d as axes3d
import math
import sympy as sp
from scipy.integrate import quad
```

```python
def f1(x):
    return np.pi* 16*x**2 - 16
ans, err = quad(f1,-1,1)
print (ans)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1, projection= '3d')
u = np.linspace(-1,1,300)
v = np.linspace(0,2*np.pi,300)
U,V = np.meshgrid(u,v)
X=U
Y1 =(U**2 +3)*np.cos(V)
Z1 =(U**2 +3)*np.sin(V)
Y2 =(5-U**2)*np.cos(V)
Z2 = (5-U**2)*np.sin(V)
ax.plot_surface(X, Y1, Z1, alpha=0.5, color='blue', rstride=6, cstride=12)
ax.plot_surface(X, Y2, Z2, alpha=0.5, color='red', rstride=6, cstride=12)
plt.show()
```
